```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Iris.csv')
```

```python
print('Features and their types')
df.dtypes
```

```python
# Create histograms for each numeric feature
df.drop(columns='Species').hist(figsize=(10, 8), bins=20)
plt.suptitle("Histograms for Each Feature")
plt.show()
```

```python
# Create boxplots for each feature
plt.figure(figsize=(10, 8))

# Plot boxplots for each feature
sns.boxplot(data=df.drop(columns='Species'), orient='h')
plt.title("Boxplots for Each Feature")
plt.show()


```

```python
# Compare distributions and identify outliers.

sns.boxplot(x='SepalWidthCm', data=df)
plt.show()
sns.boxplot(x='SepalLengthCm', data=df)
plt.show()
sns.boxplot(x='PetalWidthCm', data=df)
plt.show()
sns.boxplot(x='PetalLengthCm', data=df)
plt.show()
```
