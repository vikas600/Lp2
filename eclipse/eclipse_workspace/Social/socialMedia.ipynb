```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
data = pd.read_csv('Social_Network_Ads.csv')  # make sure the file is in your working directory
print(data.head())  # optional: check first few rows
```

```python
# Assuming dataset has 'Age', 'EstimatedSalary' as features and 'Purchased' as target
X = data[['Age', 'EstimatedSalary']]
y = data['Purchased']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)  
```

```python
TN, FP, FN, TP = cm.ravel()
print(f"TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}")
```

```python
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print(f"Error Rate: {error_rate:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
# Metric	Formula	Meaning
# Accuracy	(TP + TN) / Total	Kitna sahi prediction
# Error Rate	1 - Accuracy	Kitna galat prediction
# Precision	TP / (TP + FP)	Jab predict kiya 1, tab kitna sahi tha
# Recall	TP / (TP + FN)	Total actual 1 mein kitna sahi pakda
```
