```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Titanic dataset load
titanic = sns.load_dataset('titanic')

# Dataset ka ek chhota preview dekh lo
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
sns.countplot(x='survived', data=titanic)
plt.title("Count of Survivors vs Non-Survivors")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Plotting the count of passengers by class
sns.countplot(x='pclass', data=titanic)
plt.title("Count of Passengers by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Plotting the relationship between age and survival status
sns.boxplot(x='survived', y='age', data=titanic)
plt.title("Age Distribution with Respect to Survival Status")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
# 4. Visualizing the relationship between fare and survival status
sns.boxplot(x='survived', y='fare', data=titanic)
plt.title("Fare Distribution with Respect to Survival Status")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
```

```python
# Plot histogram for the distribution of fare
sns.histplot(data=titanic, x='fare', kde=True)  # kde=True adds a Kernel Density Estimation curve
plt.title("Distribution of Titanic Ticket Fare")
plt.xlabel("Fare Price")
plt.ylabel("Number of Passengers")
plt.show()
```
